# Dproxy

分布式代理服务，主要提供动态代理功能，添加多条ADSL拨号线路，实现多地区切换与持续服务。
```




                                        VPS [codertool.cn]

      ADSL VPS                                                                   ADSL VPS
                                       +------------------+
+------------------+                   |   Ngrok Server   |                 +------------------+
|   Ngrok Client   +-------------------+                  +-----------------+   Ngrok CLient   |
+------------------+                   +------------------+                 +------------------+
+------------------+                   +------------------+                 +------------------+
|Shadowsocks Server|                +--+Shadowsocks Client|                 |Shadowsocks Client|
+------------------+                |  +------------------+                 +------------------+
                                    |  +------------------+
                                    |  |     Delegate     +--+
                                    |  +------------------+  |
                                    v                        v
                                Socket5 Proxy           HTTP Proxy

```


> 目前公网VPS为 codertool.cn, Socket5代理端口为 6188， HTTP代理端口为 888，无认证


## 主要模块

1. 基于Shadowsocks开发代理服务器

2. 使用Ngrok完成VPS穿透，后续会基于此开发服务配置功能，实现无需手动配置服务的端口等

3. 使用Delegate 代理转换

4. 公网VPS

## 实现

### 代理服务

代理服务，在该项目中，代理服务分为服务端（shadowsocks-server）和客户端（shadowsocks-local）。
服务端部署在每条ADSL线路的机器上，开启代理服务端并借助后诉的ngrok实现对外提供代理功能;
客户端部署在能被使用者访问到的机器上，负责开启Socket代理服务供使用者连接，并选择与部署在ADSL拨号线路VPS上的服务端建立连接，
完成的多出口选择。

其中对原有Shadowsocks程序，进行了更改，增加了命令请求协议，用于客户端与服务端的控制通信，现版本实现心跳监测功能，
用于客户端对服务端服务的提供与否进行检测，提高客户端选择有效线路的命中率。

### Ngrok穿透服务

由于ADSL拨号VPS没有公网IP，其提供的Shadowsocks服务无法被公网访问到，所以在每台拨号VPS都部署Ngrok客户端，连接部署在公网VPS上的Ngrok服务端；
实际使用时Shadowsocks客户端连接的Shadowsocks服务端时是与本地某个端口通信，该端口被Ngrok服务端转发至拨号VPS上，由Ngrok客户端转发至Shadowsocks服务端。

### Delegate

Delegate负责将Shadowsocks客户端开启的Socket代理转换至HTTP代理。

## 现状

1. 拨号由部署在拨号VPS上的程序定时执行，无法统一调度

2. Delegate可能会存在请求性能下降，如果可以直接使用Socket5代理性能最好

3. Ngrok没有Web控制界面，所有端口的确定与更改需要手动炒作配置文件

4. Ngrok存在性能问题，掉线恢复延时较高（数秒）

5. 目前拨号VPS上部署有抓取程序，拨号频率较高，请求会有10%的超时 [使用程序可增加重试次数]
